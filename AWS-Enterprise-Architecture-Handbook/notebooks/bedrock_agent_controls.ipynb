# Bedrock Agent Control Lab


## Scenario

Define a bounded agent workflow for security triage.


## Tool Policy

Model allowed and approval-required tools.


```python
tools = {
  "get_finding": "read",
  "query_cmdb": "read",
  "open_ticket": "write",
  "isolate_host": "approval_required"
}
tools
```

## Decision Gate

Require human approval before consequential containment.


```python
def may_execute(tool, approved=False):
    return tools[tool] != "approval_required" or approved
may_execute("isolate_host", approved=False)
```

Log prompts, tool calls, sources, approvals, and outcomes.
