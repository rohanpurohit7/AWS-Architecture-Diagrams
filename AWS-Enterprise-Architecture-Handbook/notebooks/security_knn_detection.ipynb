# Security Analytics KNN Lab


## Scenario

Model network events as feature vectors for anomaly exploration.


## Synthetic Data

Generate a safe synthetic dataset.


```python
import numpy as np
rng = np.random.default_rng(42)
normal = rng.normal([50, 200, 1], [10, 30, .2], size=(500,3))
anomaly = rng.normal([200, 900, 5], [30, 100, .8], size=(20,3))
X = np.vstack([normal, anomaly])
X.shape
```

## KNN Distance

Use neighbor distance as one anomaly signal.


```python
from sklearn.neighbors import NearestNeighbors
nn = NearestNeighbors(n_neighbors=8).fit(X)
distances, _ = nn.kneighbors(X)
score = distances[:,-1]
score[-10:]
```

## Interpretation

KNN supports analyst triage; it does not prove maliciousness.
